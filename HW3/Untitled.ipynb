```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mu(m,s):
    return np.exp(m+s*s/2)
def sigma(m,s):
    return (np.exp(m+s*s/2))**2*(np.exp(s*s)-1)

def fun1(m,s):
    return sigma(m,s)*mu(m,s)/(2*sigma(m,s)**2+mu(m,s)**2)
def fun2(m,s):
    return (mu(m,s)**3-2*sigma(m,s)**2*mu(m,s))/((2*sigma(m,s)**2+mu(m,s)**2)**2)
```

```python
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')


x = np.linspace(-10,10,1000)

plt.plot(x,fun1(x,0.1),color='b',label='s = 0.1')

plt.plot(x,fun1(x,1),color='r',label='s = 1')

plt.plot(x,fun1(x,2),color='y',label='s = 2')

plt.legend(loc='upper right')

plt.xlabel('m')
plt.ylabel('first derivative')

plt.show()
```

```python
plt.figure(num=None, figsize=(12, 8), dpi=80, facecolor='w', edgecolor='k')

x = np.linspace(-10,10,1000)

plt.plot(x,fun2(x,0.1),color='b',label='s = 0.1')

plt.plot(x,fun2(x,1),color='r',label='s = 1')

plt.plot(x,fun2(x,2),color='y',label='s = 2')

#plt.axvline(x=np.log(1.5))

plt.legend(loc='upper right')

plt.xlabel('m')
plt.ylabel('second derivative')

plt.show()
```
